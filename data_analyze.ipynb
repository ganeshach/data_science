```python
import pandas as pd
from pathlib import Path

file_path = Path("all_samples_data.csv")
data = pd.read_csv(file_path)

print(data)
```

```python
mean_values = data.mean()
variance_values = data.var()

print("Mean:")
print(mean_values)

print("\nVariance:")
print(variance_values)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_csv("all_samples_data.csv")
sample_means = data.mean()

bins = np.arange(sample_means.min(), sample_means.max() + 0.2, 0.2)

plt.figure(figsize=(8, 5))
plt.hist(sample_means, bins=bins, edgecolor="black")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.title("Histogram of Sample Means")
plt.show()
```
